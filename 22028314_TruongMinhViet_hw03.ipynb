```python
before_2000 = imdb.where("Year", are.below(2000))
after_or_in_2000 = imdb.where("Year", are.above_or_equal_to(2000))

before_2000_avg = np.average(before_2000.column("Rating"))
after_or_in_2000_avg = np.average(after_or_in_2000.column("Rating"))

print("Average before 2000 rating:", before_2000_avg)
print("Average after or in 2000 rating:", after_or_in_2000_avg)
```
